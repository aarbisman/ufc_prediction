### Starting with the all-events UFC page, get all event links

```python
all_events_url="http://ufcstats.com/statistics/events/completed?page=all"

all_events_soup = make_soup(all_events_url)
all_event_links = get_all_event_details(all_events_soup)
print(len(all_event_links))
```

```python
event_set = all_event_links[740:]

for i in range(len(event_set)):
    if i % 10 == 0:
        print(i)

    event = event_set[i]
    event_id = event.split('/')[-1]

    scraped_dat = get_event_data_from_links([event], totals_table_structure, sig_str_table_structure, fight_results_table_structure)

    file_name = 'data/all_events_seperated/' + event_id + '.json'
    with open(file_name, 'w') as fp:
        json.dump(scraped_dat, fp, indent=4)
```
